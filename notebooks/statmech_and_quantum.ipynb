# Statistical mechanics and quantum dynamics

This notebook uses the two Monte Carlo temperature sweep paths and a
1D quantum tunneling problem to show the breadth of the package.

```python
import jax
import jax.numpy as jnp
import neurosim as ns

jax.config.update("jax_enable_x64", True)
```

```python
lattice = ns.IsingLattice(size=(16, 16), J=1.0)
temperatures = jnp.linspace(1.5, 3.5, 9)
ising = ns.sweep_temperatures(
    lattice,
    temperatures,
    n_sweeps=40,
    n_warmup=80,
    algorithm="wolff_cluster",
    key=jax.random.PRNGKey(0),
)
ising.magnetizations.shape, ising.susceptibilities.shape
```

```python
barrier = ns.SquareBarrier(height=5.0, width=1.0, center=10.0)
psi0 = ns.GaussianWavepacket(x0=5.0, k0=3.0, sigma=0.5)
quantum = ns.solve_schrodinger(
    psi0=psi0,
    potential=barrier,
    x_range=(-5.0, 25.0),
    t_span=(0.0, 5.0),
    n_points=600,
    dt=0.01,
    save_every=20,
)
quantum.transmission_coefficient
```

The temperature sweep and tunneling solver are both deterministic
enough to be good notebook examples without introducing extra
dependencies.
