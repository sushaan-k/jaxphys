# Electromagnetism and optics

This notebook combines the scalar diffraction helpers with the 2D
FDTD grid. The goal is to show both the analytic and grid-based sides
of the package in one place.

```python
import jax
import jax.numpy as jnp
import neurosim as ns

jax.config.update("jax_enable_x64", True)
```

```python
single = ns.single_slit(slit_width=1e-4, wavelength=500e-9, n_points=800)
double = ns.double_slit(slit_width=5e-5, slit_separation=2e-4, wavelength=500e-9, n_points=800)
airy = ns.circular_aperture(diameter=1e-3, wavelength=500e-9, n_points=800)
float(jnp.max(single.intensity)), float(jnp.max(double.intensity)), float(jnp.max(airy.intensity))
```

```python
grid = ns.EMGrid(size=(60, 60), resolution=0.01, pml_layers=8)
grid.add_source(ns.PlaneWave(frequency=3.0e9, y=6, amplitude=1.0))
grid.add_conductor(ns.Wall(y=30, gap_start=26, gap_end=34))
fields = grid.simulate(t_span=(0.0, 2.0e-10), save_every=5)
fields.ez.shape, fields.t.shape
```

A simple FDTD screen slice is enough for a smoke test. The notebook is
not pretending to be a production optics package; it is showing the
actual APIs that exist in the repo today.
